#### It is often necessary to write a custom function

The basic syntax is:

```python
def name_of_function(arguments_separated_by_commas):
    code
    code
    return values_separated_by_commas
```
    - def begins the function definition
    - The function name should describe what the function does
    - Arguments are listed inside the parentheses and separated by commas
    - Indentation is required for all code inside the function
    - A return statement sends values back to the caller
    - A function can include a docstring immediately after the definition. The docstring is enclosed in triple quotes (""") and is displayed when help() is called

Example from the kaggle tutorial:

### Here is an example from the kaggle tutorial

````python
def least_difference(a, b, c):
    """Return the smallest difference between any two numbers
    among a, b and c.
    
    least_difference(1, 5, -5)
    4
    """
    diff1 = abs(a - b)
    diff2 = abs(b - c)
    diff3 = abs(a - c)
    return min(diff1, diff2, diff3)

least_difference(10, 20, 30)
````

````python
help(least_difference)
````

My own examples:

````python
def apply_discount(cost, discount):
    """Return the cost for an item after a discount is applied
    accepts a % as input for discount

    >>>price_after_discount(100,"10%")
    $90.00
    """
    discount_rate = float(discount.rstrip("%")) / 100
    return cost * (1 - discount_rate)

price=(apply_discount(93,"10%"))
print(f"Discounted price: ${price:.2f}")
````

- I used printing with formatting to try and avoid any errors with rounding bias (round half to even).  However, AI tells me that even using format to display only two decimals can occasionally have a rounding bias. Functions can return a lot of variable types.  I have not covered all the variable types yet.  But in addtion to float and int, functions can return
- Strings
- Bools
- Lists
Here are examples of each:

````python
# String
def full_name(first, last):
    """Return a person's full name."""
    return first + " " + last
print(full_name("Arthur", "Dent"))
````

````python
# Bool
def is_even(number):
    """Return True if the number is even."""
    return number % 2 == 0
print (is_even(3))
print (is_even(4))
````

````python
# Lists
def get_colors():
    return ["red", "green", "blue"]

print(get_colors())
````

You can supply functions as arguments to other functions - these are called higher order functions

````python
def mult_by_five(x):
    return 5 * x

def call(fn, arg):
    """Call fn on arg"""
    return fn(arg)

def squared_call(fn, arg):
    """Call fn on the result of calling fn on arg"""
    return fn(fn(arg))

print(
    call(mult_by_five, 1),
    squared_call(mult_by_five, 1), 
    sep='\n', # '\n' is the newline character - it starts a new line
)
````
