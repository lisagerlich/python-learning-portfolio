# A Notebook Introducing List Comprehensions

List comprehensions are a Python feature that allows many simple `for` loops to be written in a more compact form.

They are most commonly used when creating a new list from existing data.

You can use them to:
- create new lists
- filter items
- transform data
- extract information
- perform conditional transformations

List comprehensions can be overused.  If the task is complex, a standard for loop is often easier to read and maintain.

##### Create a list of square numbers from a range

```python
squares = [n**2 for n in range(10)]
print(squares)
```

Here is the for loop that does the same thing:

```python
squares = []
for n in range(10):
    squares.append(n**2)
print(squares)
```

I think that list comprehensions are easier to read and write, if you use white space.

```python
squares = [
    n**2 # What gets added to the list
    for n in range(10) # The values we are iterating through to create the list
]
print(squares)
```

##### Take an existing list and filter for items that meet a condition

```python
evens = [
    square
    for square in squares
    if square % 2 == 0
]
print(evens)
```

##### Transform a list of strings to lowercase

```python
coffeebrands = ['Folgers', 'Maxwell House', 'Gevalia', "Chock Full O' Nuts"]
lc_coffeebrands = [
    coffeebrand.lower()
    for coffeebrand in coffeebrands
]
print(lc_coffeebrands)
```

####  Build a new list by extracting data from items in an existing list

```python
names = [
    "Winston Smith",
    "George Smiley",
    "Anna Karenina"
]
firstnames = [
    name.split()[0]
    for name in names
]
print(firstnames)
```

#### Change the item based on a condition

```python
numbers = [1, 2, 3, 4, 5]

labels = [
    "Even" 
    if num % 2 == 0 
    else "Odd" 
    for num in numbers
]
print(labels)
```

```python
scores = [95, 72, 88, 61, 100]

results = [
    "Pass" 
    if score >= 70 
    else "Fail" 
    for score in scores
]
print(results)
```
