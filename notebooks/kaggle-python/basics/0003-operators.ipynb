# Python Arithmetic Operators and Order of Operations


## Python uses familiar arithmetic operators

- Addition

```python
print(911 + 411)
```

- Subtraction

```python
print(911 - 411)
```

- Multiplication

```python
print(911 * 411)
```

- True Division

```python
print(42 / 7)
print(42 / 8)
print(42 / 9)
```

- Floor Division:  returns the integer part of a decimal result

```python
print(42 // 8)
print(42 // 9)
```

- Modulus: returns the remainder after division
- Example: 42 % 5 returns 2 because 42 = 5 * 8 + 2

```python
print(5 * 8 + 2)
print(42 % 5)
```

- Exponentiation

```python
print(10 ** 2)
print(10 ** 3)

# Get the square root - returned as a float
print(25 ** .5)
print(100 ** .5)
```

* Negation

```python
g = 42
print(-g)
```

##### Among the operators shown here, all return an integer when both operands are integers, except /, which always returns a float. Using ** with a fractional exponent (such as .5) also returns a float.

```python
print(type(100 + 2))
print(type(100 - 2))
print(type(100 * 2))
print(type(100 / 2))
print(type(100 // 2))
print(type(100 % 2))
print(type(100 ** 2))
print(type(100 ** .5))
print(type(100.0 + 2))
```

## Python follows the **PEMDAS** order of operations: **P**arentheses, **E**xponents, **M**ultiplication/**D**ivision, **A**ddition/**S**ubtraction.

- Parentheses First
    -  4 * (5 + 3)
    -  4 * 8
    -  32

```python
result = 4 * (5 + 3)
print(result)
```

- Exponents Next
    - 5 * 2
    - 5 * 4
    - 20

```python
result = 5 * 2 ** 2
print(result)
```

- Multiplication and Division (includes Floor Division and Modulo) Before Addition and Subtraction
    - 2 + 5 * 3
    - 2 + 15
    - 17

    - 4 + 20 // 3
    - 4 + 6
    - 10

    - 4 + 20 % 3
    - 4 + 2
    - 6

```python
result = 2 + 5 * 3
print(result)
result = 4 + 20 // 3
print(result)
result = 4 + 20 % 3
print(result)
```

- Divide and Multiply rank equally
- Addition and Subraction rank equally
- When operations rank equally, calculate from left to right
    - 30/5 * 3
    - 6 * 3
    - 18.0

```python
result = 30/5 * 3
print(result)
```

##### Here is a complicated example of PEMDAS
    - ((7 + 5) ** 2 % 13) + (50 // (3 + 2)) * 2 - 4 ** 2
    - (12 ** 2 % 13) + (50 // 5) * 2 - 4 ** 2
    - (144 % 13) + (10 * 2 - 4 ** 2)
    - 1 + 20 - 16
    - 21 - 16
    = 5

```python
print(((7 + 5) ** 2 % 13) + (50 // (3 + 2)) * 2 - 4 ** 2)
```

    -  (0 + 9) * 2 - 10 ** 2 // 5

    -  9 * 2 - 100 // 5
