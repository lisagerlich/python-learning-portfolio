# Python Loops
Loops allow a block of code to be executed repeatedly. In Python, loops are commonly used to iterate over collections such as lists, tuples, strings and ranges or to repeat code while a condition remains true.
Python provides two main types of loops:
1. for loops repeat once for each item in a collection.
2. while loops repeat as long as a condition remains True.

**The -for loop- specifies**
- the variable to use
- the set of values to loop over
- the word "in" links them together

Examples:

```python
# Using a list
planets = ['Mercury', 'Venus', 'Earth', 'Mars', 'Jupiter', 'Saturn', 'Uranus', 'Neptune']
for planet in planets: 
    print(planet, end=' ') # The end parameter specifies what is printed after the output. By default, print() ends with a newline (\n).
```

```python
# Using a tuple
multiplicands = (2, 2, 2, 3, 3, 5)
product = 1
for mult in multiplicands:
    product = product * mult
product
```

```python
# A string
s = 'steganograpHy is the practicE of conceaLing a file, message, image, or video within another fiLe, message, image, Or video.'
msg = ''
# print all the uppercase letters in s, one at a time
for char in s:
    if char.isupper():
        print(char, end='')        
```

##### Kaggle just introduced "range"
range() is a built-in Python function that generates a sequence of numbers. It is commonly used with for loops when you want to repeat code a specific number of times.  By default the range starts at 0.

```python
for i in range(5):
    print("All work and no play makes Jack a dull boy. i =", i)
```

```python
for i in range(1,5):
    print("The ants go marching", i, "by", i, "hurrah, hurrah.")
```

```python
for i in range(5):
    i = i + 1
    print("The ants go marching ", i, " by ", i, "hurrah, hurrah.")
```

**The -while loop- iterates until some condition is met**
- The argument of the while loop is evaluated as a boolean statement
- the loop is executed until the statement evaluates to False.

Examples:

```python
i = 0
while i < 10:
    print(i, end=' ')
    i += 1 
```

```python
i = 5
while i > 0:
    print(i)
    i -= 1
print("Blast Off!")
```
