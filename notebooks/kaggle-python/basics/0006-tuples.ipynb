# A short notebook about tuples

1. A tuple is a collection of items
2. It uses parentheses () instead of square brackets []
3. It cannot be modified after creation

````python
my_tuple = (10, 20, 30)

print(my_tuple)
print(type(my_tuple))
````

- A tuple behaves a lot like a list, except its individual items cannot be changed after the tuple is created.
- Like strings, the variable holding the tuple can still be reassigned to a different tuple.

````python
ytuple = (255, 255, 255)
xstring = "abc"

ytuple = (0, 0, 0)
xstring = "xyz"
````

- The above works because you are reassigning the variable not changing elements within it.
- The following will produce an error:
```
ytuple[0] = 255
xstring[0] = "d"
```

Lists are mutable and the elements within them can be changed.

````python
ylist = [ "apples", "oranges", "bananas" ]
print(ylist)
ylist[1] = "tangerines"
print(ylist)
````

Because tuples are immutable, they're often used for things that shouldn't change:

````python
point = (10, 20)
rgb_color = (255, 0, 0)
date = (2026, 6, 6)
````

Tuples do support indexing, just like lists

````python
mytuple = ("yuccas", "palms", "cacti", "ocotillo")
print(mytuple[2])
````
