# Booleans and Conditionals

Boolean operators produce boolean values

| Operation | Description | 
|-----------|-------------|
| a == b    | a equal to b |
| a < b     | a less than b |
| a > b     | a greater than b |
| a >= b    | a greater than or equal b |
| a <= b    | a less than or equal to b |
| a != b    | a not equal to b |



````python
def can_run_for_president(age):
    """Can someone of a given age for president of the United States"""
    return age >= 35
print("Can a 19-year-old run for president?", can_run_for_president(19))
print("Can a 45-year-old run for president?", can_run_for_president(45))
````

Sometimes comparisons may not behave the way you first expect
- if you compare an integer to a float, python will follow internal rules to convert the numbers to the same type
- if you compare a string to an integer or float, Python will not automatically convert a string to a number when using ==.

````python
3 == 3.0
````

````python
'3' == 3
````

If you want to compare the numerical output of a string, convert it to a number first

````python
int('3') == 3
````

Comparison operators can be used with arithmetic operators

````python
def is_divisible_by_10(n):
    return (n % 10) == 0
print("Is 300 divisible by 10?", is_divisible_by_10(300))
print("Is 205 divisible by 10?", is_divisible_by_10(205))
````

You can combine boolean values using the boolean operators and, or, and not.

````python
def should_window_be_closed(window,raining):
    return(window == "Open") and raining
print(should_window_be_closed("Open", True))
print(should_window_be_closed("Closed", True))
print(should_window_be_closed("Open", False))
print(should_window_be_closed("Closed", False))
````

````python
def car_engine_needs_attention(engine_light_on, grinding_noise):
    return(engine_light_on or grinding_noise)
print(car_engine_needs_attention(True, True))
print(car_engine_needs_attention(True, False))
print(car_engine_needs_attention(False, True))
print(car_engine_needs_attention(False, False))
````

````python
def should_go_for_walk(raining):
    return not raining

print(should_go_for_walk(True))
print(should_go_for_walk(False))
````

````python
def go_to_work(rain_level, is_workday):
    """
    Go only if it is not both raining and a workday.
    """
    return not (rain_level > 0 and is_workday)

print(go_to_work(rain_level = 1, is_workday = False))
print(go_to_work(rain_level = 0, is_workday = False))
print(go_to_work(rain_level = 0, is_workday = True))
print(go_to_work(rain_level = 1, is_workday = True))
````

A good way to think of the expression above is:
- not (a and b) is the same as
- (not a) or (not b)
- not the same as
- (not a) and (not b)

When writing complex boolean expressions that use `and`, `or`, and `not`, parentheses can make the code much easier to read. Although Python has rules that determine the order in which these operators are evaluated, explicit grouping helps clarify the intended logic and reduces the chance of introducing bugs. For readability, long expressions can also be split across multiple lines.

#### Booleans are most useful when combined with conditional statements, using the keywords if, elif, and else.

Conditional statements, often referred to as if-then statements, let you control what pieces of code are run based on the value of some Boolean condition.  

````python
def inspect(x):
    t = type(x)
    if t != int and t != float:
        print(x, "is unlike anything I've ever seen...")
    elif x == 0:
        print(x, "is zero")
    elif x > 0:
        print(x, "is positive")
    else:
        print(x, "is negative")
              
inspect(0)
inspect(-15)
inspect(42)
inspect(3.14)
inspect("blue")
````

- A more advanced way of stating ```if t != int and t != float:``` is ```if not isinstance(x, (int, float)):```

### The structure of a Python if statement:

```
if conditional:
    command(s) that run if the condition is True
elif conditional:
    command(s) that run if the condition is True
elif conditional:
    command(s) that run if the condition is True
else:
    command(s) that run if none of the previous conditions were True
```

An if statement begins with an if clause and may contain
zero or more elif clauses and zero or one else clause.

The end of an if statement is indicated by a line that is
less indented than the if block and does not begin with
elif or else.

````python
def f(x):
    if x > 0:
        print("Only printed when x is positive; x =", x)
    elif x < 0:
        print("Only printed when x is negative; x =", x)
    print("Always printed, regardless of x's value; x =", x)
f(1)
f(0)
f(-1)
````

#### Boolean Conversions
- int() converts decimals to ints
- float() converts ints to floats
- bool() converts ints, floats, strings, lists and tuples to bool

````python
print(int(2.0))
print(float(2))

print(bool(1)) # all numbers are treated as true, except 0
print(bool(0))
print(bool("asf")) # all strings are treated as true, except the empty string ""
print(bool(""))
print(bool([1,2,3])) # all lists are treated as true, except the empty lists []
print(bool([]))
# Generally empty sequences (strings, lists, etc) are "falsey" and the rest are "truthy"
````

Non-Boolean objects can be used in if conditions and other places where a Boolean value is expected.

- Python implicitly converts the object to either True or False.
- Empty sequences and collections are treated as False.
- Non-empty sequences and collections are treated as True.
- The number 0 is treated as False, while non-zero numbers are treated as True.

This can be useful when checking whether a sequence is empty or whether a numeric value is non-zero.

````python
# check to see if a number is nonzero
def is_nonzero(num):
    if num:
        print(num, "is nonzero")
    else:
        print(num, "is zero")
num = 1
is_nonzero(num)
num = 3.14
is_nonzero(num)
num = 0
is_nonzero(num)
num = 0.00000
is_nonzero(num)
````

````python
# Check to see if a string is empty
def check_string(string):
    if string:
        print("The string", string, "is not empty")
    else:
        print("The string is empty")
string = "A horse, a horse, my kingdom for a horse"
check_string(string)
string = ""
check_string(string)
string = "querty"
check_string(string)
````

````python
# check to see if a list is empty
def has_items(items):
    if items:
        print("The list", items, "is not empty")
    else:
        print("The list is empty")
items = [ 3, 4, "apples" ]
has_items(items)
items = ["Fiats" , "Ladas", "Jeeps" ]
has_items(items)
items = []
has_items(items)
````
