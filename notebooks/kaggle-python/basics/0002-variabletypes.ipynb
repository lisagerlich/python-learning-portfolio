# Strings

```python
# str (string txt)
s = "hello"
print(s)
# strings are immutable and, therefore, cannot be modified directly
# s[0] = "j"

# to change s, do one of two things
# 1
s = "jello"
print(s)
# python created a new string "jello".  The variable s is reassigned to point to the new string
# the old string no longer has a variable pointong to it and will eventually be removed by a garbage collection process that runs in the background.

# 2
# the same thing happens when you use a method on a string
s = s.upper()
print(s)
# a new string, "JELLO" is created.
```

# Decimals

```python
# float
y = 3.14
print(y)
print(type(y))
```

- use round() to round() decimals 
- round() returns a new value the old value remains the same
- round() uses bankers rounding even numbers followed by a 5 will be rounded down odd numbers followed by a 5 will be rounded up
- python uses bankers rounding to Python's behaviorto avoid bias when large amounts of data are rounded repeatedly.


```python
#You can use round() to round decimals
w = round(2.5)
x = round(3.2)
y = round(3.8)
z = round(3.5)

print(w,x,y,z)
```

- you can round to tens, hundreds, etc.

```python
print(round(1234, -1))
print(round(1234, -2))
print(round(1234, -3))
```

- You can specify which decimal point to round on

```python
x = 3.14159

print(round(x, 2))
print(round(x, 3))
print(round(x, 4))
```

# bool (Booleans)
- Have a value of True (1) or False (0)

```python
# r will be an int with a value of 2
r = True + True
print(r)
print(type(r))

# r will be an bool with a value of True
r = 5 > 3
print(r)
print(type(r))

# r will be a bool with a value of False
r = 2 + 2 == 5
print(r)
print(type(r))
```

- bools are used in if statements

```python
is_raining = True
if is_raining:
    print("Bring an umbrella")
else:
    print("Leave your umbrella at home")
```

- bools can be used to check to see if a list is empty or not 

```python
nums = []
if nums:
    print("Has items")
else:
    print("Empty list")
    
nums = [1, 2, 3]
if nums:
    print("Has items")
else:
    print("Empty list")
```

- Functions can also return bools

```python
# Will return bool with value of True
def is_even(n):
    return n % 2 == 0
print(is_even(4))
print(type(is_even(4)))
```

 - Bools are used with logic operators ("and","or","not")

```python
# Will return a bool with a value of True
age = 25
has_ticket = True

can_enter = age >= 18 and has_ticket
print(can_enter)
print(type(can_enter))
```

# Lists
-Lists are mutable ordered collections of items in parenthesis separated by commas
- Use methods to manipulate lists
- Most list methods mutate the existing list instead of creating a new one.

```python
nums = [1, 2, 3]
print(nums)
print(type(nums))
```

append(): add an item to a list

```python
nums = [1, 2, 3]
nums.append(4)
print(nums)
```

remove() remove the first matching item from a list

```python
# Will remove the first 2 from the list
nums = [1, 2, 3, 2]
nums.remove(2)
print(nums)
```

pop() Removes and returns an item from a list.  By default it will remove and return the last item.

```python
nums = [10, 20, 30]
x = nums.pop()
print(x)
print(type(x))
print(nums)
```

pop(int):selects and removes a specific item from a list.  Lists are indexed starting at 0.

```python
# Will remove and return the first item in the list
x = nums.pop(0)
print(x)
print(nums)
```

```python
# Will remove and return the 6th item in the in the list
nums = [ 100, 200, 300, 400, 500, 600, 700 ]
x = nums.pop(5)
print(x)
print(nums)
```

```python
# Will remove and return the 3rd item in the in the list
x = nums.pop(2)
print(x)
print(nums)
```

sort(): Sorts the list in place

```python
nums = [ 20, 10, 5, 26, 7, 8, 9, 10, 11 ]
nums.sort()
print(nums)

# prove mutability - the following will return 9 and remove 9 from the list
x = nums.pop(3)
print(x)
print(nums)
```

reverse(): reverses the list in place

```python
nums = [ 20, 10, 5, 26, 7, 8, 9, 10, 11 ]
nums.reverse()
print(nums)
```

len(): returns the length of the list

```python
# Will return 9
x = len(nums)
print(x)
print(type(x))
```

```python
# but the last item in the list will be indexed at 8, because lists are indexed starting at '0'
x = nums[8]
print(x)
# List will not be changed
print(nums)
```
