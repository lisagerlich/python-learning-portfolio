# Objects and Methods

Python is an object-oriented language, which means that almost everything you work with is an object.
`42
"Hello"
[1, 2, 3]
True`
- Objects are pieces of data and can have things attached to them.
- Methods are functions attached to an object and are used to perform actions.
- Attributes store information about an object.
- Methods and attributes are accessed using a dot notation.
- Different object types have different methods and attributes.

#### This notebook focuses on methods because they are commonly used in beginner Python code.

Integers are objects and have methods, but unlike strings and lists, integer methods are used less frequently. Most work with integers is done using operators such as +, -, *, /, %, and **.  I am not going to worry about methods for integers and decimals until I need them.  I have already been using methods with strings and lists.  

```python
mylist = [ 'no mama', 'no papa' ]

for text in mylist:
    print(text)
help(mylist.append)
```

```python
mylist.append( 'no whisky soda' )
for text in mylist:
    print(text)
print()
help(mylist.append)
```

help() is most useful for functions and methods. Attributes are usually explored by viewing their value or checking their type.

```python
text = "zaphod beeblebrox"
text = text.upper()
print(text)
print()
help(text.upper)
```
