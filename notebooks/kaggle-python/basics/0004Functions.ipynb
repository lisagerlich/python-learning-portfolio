##### Help is one of the most useful builtin functions
- help() displays two things:
    1. The header of the function and the arguments that it takes
    2. A brief English description of what the function does

````python
help(print)
````

#### It is often necessary to write a custom function

The basic syntax is:

```python
def name_of_function(arguments_separated_by_commas):
    code
    code
    return values_separated_by_commas
```
    - def begins the function definition
    - The function name should describe what the function does
    - Arguments are listed inside the parentheses and separated by commas
    - Indentation is required for all code inside the function
    - A return statement sends values back to the caller
    - A function can include a docstring immediately after the definition. The docstring is enclosed in triple quotes (""") and is displayed when help() is called

Example from the kaggle tutorial:

### Here is an example from the kaggle tutorial

````python
def least_difference(a, b, c):
    """Return the smallest difference between any two numbers
    among a, b and c.
    
    least_difference(1, 5, -5)
    4
    """
    diff1 = abs(a - b)
    diff2 = abs(b - c)
    diff3 = abs(a - c)
    return min(diff1, diff2, diff3)

least_difference(10, 20, 30)
````

````python
help(least_difference)
````

My own examples:

````python
def apply_discount(cost, discount):
    """Return the cost for an item after a discount is applied
    accepts a % as input for discount

    >>>price_after_discount(100,"10%")
    $90.00
    """
    discount_rate = float(discount.rstrip("%")) / 100
    return cost * (1 - discount_rate)

price=(apply_discount(93,"10%"))
print(f"Discounted price: ${price:.2f}")
````

A more basic example of the function above

````python
def apply_discount(cost, percent):
    return cost * (1 - percent / 100)
    
print(apply_discount(100,8))
````

#### What I learned
    - Functions can accept multiple arguments
    - A function can return a calculated value
    - String methods can be used to clean input before processing
    - The float() function converts a string to a numeric value
    - rstrip() is a string method that removes characters from the right side (the end) of a string
    - lstrip() is a string method that removes characters from the left side (the end) of a string
    - strip() is a string method that removes characters from both ends of a string
    - a little bit about print formatting, because I wanted the return to have exactly 2 numbers after the decimal
        - the f before the opening quote tells python that this is formatted string
        - the {} tells python to take the literal value from the variable price
        - the :.2f tells python to format price as a floating-point number with 2 digits after the decimal point
