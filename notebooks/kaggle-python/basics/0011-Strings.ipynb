# Strings - a Review and a Deeper Dive

- Strings immutable objects are created by putting text between double or single quotes.

```python
#The following strings are equivilent
x = 'Praise God from Whom All Blessings Flow'
y = "Praise God from Whom All Blessings Flow"
result = x == y
print(result)
```

- You can put a single quote inside double quotes.
- The opposite also is true you can put a double quote inside single quotes.

```python
mystring = "Mr O'Brien walks every day."
print(mystring)
mystring = 'One of my favorite movies is "Dr Strangelove."'
print(mystring)
```

- You can also escape problematic characters with a backslash

```python
mystring = 'Mrs O\'Brien enjoys gardening.'
print(mystring)
mystring = "That's \"cool\""
print(mystring)
mystring = "Look, a mountain: /\\"
print(mystring)
```

- You can use a backslash to add a newline character.

```python
hello = "hello\nworld"
print(hello)
```

- Triple quotes are very useful.  
- Use them to include quotes without escaping.
- use them to create multline strings by adding an invisible newline character.

```python
s = """She said "Hello"."""
print(s)
t = """Line 1
Line 2
Line 3"""
print(t)
for i, ch in enumerate(t):
    if i in (6,13):
        print(i, repr(ch))
```

- Python also has triple quote syntax which allows you to enter newlines by hitting the return key rather than adding a newline character.

```python
triplequotedhello = """hello
world"""
print(triplequotedhello)
result = triplequotedhello == hello
print(result)
```

- The print() function automatically adds a newline character unless we specify a value for the keyword argument end other than the default value of '\n'.

```python
print("hello")
print("world")
print("hello", end='')
print("pluto", end='')
```

Strings can be thought of as sequences of characters. Almost everything we've seen that we can do to a list, we can also do to a string.
- Indexing
- Slicing
- Functions

```python
tree = 'Aspen'
print(tree[0])
print(tree[-3:])
print(len(tree))
```

Strings are iterable, so you can loop through them

```python
for char in tree:
    print(char.upper() + '!')
```

If you use a list comprehension for the loop, you will make a list out of string.

```python
letters=([ char.upper()+'!' for char in tree ])
x = type(letters)
print(x)
print(letters)
```

Unlike lists strings are immutable and do not support list methods.

```python
mylist = ['P','l','u','t','o']
mystring = "Pluto"
```

```python
mylist[0] = 'B'
print(mylist)
try: 
    mystring[0] = 'B'
except TypeError as error:
    print(error)
```

```python
mylist.append('s')
print(mylist)
try:
    mystring.append('s')
except AttributeError as error:
    print(error)
```

The empty string has length zero. The empty string the only string that Python considers as False when converting to boolean.

```python
print(bool(""))
print(bool("hello"))
```

### Like `list`, the type `str` has a lot of useful methods

```python
# ALL CAPS
claim = "Pluto is a planet!"
print(claim.upper())
# all lowercase
print(claim.lower())
# Searching for the first index of a substring
print(claim.index('plan'))
# True
print(claim.startswith('Pluto'))
```

```python
planet = 'Pluto'
claim = "Pluto is a planet!"
print(claim.startswith(planet))
print(claim.startswith("is"))
# false because of missing exclamation mark
print(claim.endswith('planet'))
print(claim.endswith('planet!'))
print(claim.endswith(planet))
```

`str.split()` breaks a string into a list of smaller strings. By default it splits on whitespace, but you can specify a separator string to split on something else.

```python
words = claim.split()
print(words)
datestr = '06-10-70'
year, month, day = datestr.split('-')
print(year)
print(month)
print(day)
datelist = datestr.split('-')
print(datelist)
```

`str.join()`, combines a list of strings into a string inserting the string before `.join()` between each item.

```python
datelist = "/".join([month, day, year])
print(datelist)
mystring = "".join("hello")
print(mystring)
mystring = "-".join("hello")
print(mystring)
```

Unicode characters can be added to strings

```python
sentence = 'What a beautiful day'
words = sentence.split()
funsentence = ' 🌅 '.join([word.upper() for word in words])
print(funsentence)
```

## String Formatting

- you can concatenate strings with `+`
- if you use an int with the plus method you will need to change it to a string first

```python
planet = 'Pluto'
position = 9
print(planet + ', we miss you.')
```

```python
try:
   print(planet + ", you'll always be the " + position + "th planet to me.")
except TypeError as err:
    print("Error:", err)
```

```python
print(planet + ", you'll always be the " + str(position) + "th planet to me.")
```

- string format can simplify string concatenation
- the format method is called on string
- the arguments to format are the objects we want to print
- `{}` are placeholders for the arguments

```python
print("{}, you'll always be the {}th planet to me.".format(planet, position))
```

Here is example that shows format being used to print:
- 2 significant digits.
- 3 significant digits.
- a percentage.
- separation with commas.

```python
pluto_mass = 1.303 * 10**22
earth_mass = 5.9722 * 10**24
population = 52910390
#         2 decimal points   3 decimal points, format as percent     separate with commas
print("{} weighs about {:.2} kilograms ({:.3%} of Earth's mass). It is home to {:,} Plutonians.".format(
    planet, pluto_mass, pluto_mass / earth_mass, population,
))
```

- You can refer to arguments with indexes to repeat a word or number.
- Triple quotes allow a string to span multiple lines.

```python
s = """Pluto's a {0}.
No, it's a {1}.
{0}!
{1}!""".format('planet', 'dwarf planet')
print(s)
```

This was just an introduction to `str.format`.  More information can be found at https://pyformat.info and https://docs.python.org/3/library/string.html#formatstrings

I was introduced to str.format() because it is still common in existing code and documentation. However, for new code I plan to use f-strings because they are more readable, concise and easier to use.  Here are the examples used in format modified to use f strings instead.

```python
print(f"{planet} you'll always be the {position}th planet to me.")
```

```python
print(f"{planet} weighs about {pluto_mass:.2e} kilograms ({pluto_mass / earth_mass:.3%} of earth's mass). It is home to {population:,} Plutonians.")
```

You can call string methods directly inside an f-string.

```python
planet_type = "planet"
new_type = "dwarf planet"
s = f"""Pluto's a {planet_type}.
No, it's a {new_type}.
{planet_type.capitalize()}!
{new_type.capitalize()}!"""
print(s)
```
